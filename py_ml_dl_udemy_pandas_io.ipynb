```python
df = pd.read_csv('/Users/ada4sh/Documents/Study/Python/Python for ML and DScience/UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/Sales_Funnel_CRM.csv')
```

```python
df
```

```python
liscenses = df[['Company','Product','Licenses']]
```

```python
#using pivot 
pd.pivot(data = liscenses, index = 'Company',columns = 'Product',values = 'Licenses')
```

```python
#using pivot Table 
pd.pivot_table(df,index='Company',columns = 'Product',values = 'Licenses',aggfunc = 'sum')
```

```python
#using groupby 
liscenses.groupby(['Company','Product']).sum()
```

```python
pd.pivot_table(df,index =['Account Manager','Contact'],values=['Licenses','Sale Price'], aggfunc='sum',fill_value=0,margins=True)
```
